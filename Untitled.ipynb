```python
from tqdm import tqdm_notebook

import pandas as pd
pd.set_option('display.max_columns', 500)

import numpy as np
from scipy.stats import gaussian_kde

import matplotlib.pyplot as plt
%matplotlib inline

from standardiser import standardise
import molvs

from rdkit import Chem
from rdkit.Chem import AllChem, MolFromSmiles, MolToSmiles, MolToInchi, InchiToInchiKey, PandasTools
from rdkit.Chem.Draw import IPythonConsole
import rdkit.Chem.Descriptors as D
# from rdkit.Chem.Descriptors import MolWt
```

```python
receptors = 'ERa,ERb,GR,MR,PR,AR'.split(',')
receptors_chembl_ids = 'CHEMBL206,CHEMBL242,CHEMBL2034,CHEMBL1994,CHEMBL208,CHEMBL1871'.split(',')
chembl_id_to_name = {a:b for a,b in zip(receptors_chembl_ids, receptors)}
```

```python
df_union = pd.DataFrame()
for receptor in receptors:
    df = pd.read_csv('primary_data/CHEMBL25-{}.csv'.format(receptor))
    df_union = pd.concat([df_union, df])
    del df
df_union.rename(lambda x:x.replace(' ','_'), axis='columns', inplace=True)
```

```python
df_clean = df_union.query('Standard_Units == "nM" and Standard_Relation == "=" and not Canonical_Smiles.isnull()')
df_clean["receptor_name"] = df_clean.Target.apply(lambda x:chembl_id_to_name[x])
```

```python
df_molecules = df_clean[["Molecule", "Canonical_Smiles"]] \
    .drop_duplicates() \
    .set_index('Molecule')
```

```python
# standardise molecules and group activities by the standardised parent molecule
class CustomStandardiser:
    def __init__(self):
        self.standardiser = molvs.Standardizer()
        self.tautomer_canonicalizer = molvs.tautomer.TautomerCanonicalizer()
        self.reionzer = molvs.charge.Reionizer()

    def standardise(self, smiles):
        mol = MolFromSmiles(smiles)
        try:
            standard_parent = standardise.run(mol) # Francis Atkinson
            standard_parent2 = self.standardiser.standardize(standard_parent) # MolVS
            tautomer_parent = self.tautomer_canonicalizer.canonicalize(standard_parent2) # MolVS tautomer
            reionized_parent = self.reionzer.reionize(tautomer_parent) # MolVS reionize
            
            return reionized_parent
        except Exception as e:
            print(e)
            display(mol)
            print('---')
            return None
```

```python
cs = CustomStandardiser()
```

```python
std_mols = np.asarray(list(map(cs.standardise, tqdm_notebook(df_molecules.Canonical_Smiles.values))))
```

```python
np.save('secondary_data/std_mols.npy',std_mols)
```

```python
std_mols = np.load('secondary_data/std_mols.npy', allow_pickle=True)
```

```python
df_molecules['std_smiles'] = [x if x is None else MolToSmiles(x) for x in std_mols]
df_molecules['std_inchi']  = [x if x is None else MolToInchi(x) for x in std_mols]
df_molecules['std_inchi_key'] = [x if x is None else InchiToInchiKey(x) for x in df_molecules.std_inchi]
```

```python
df_molecules_std = df_molecules[std_mols != None] \
    .drop_duplicates('std_inchi_key') \
    .set_index('std_inchi_key') \
    .drop('Canonical_Smiles', axis='columns')
```

```python
df_molecules.drop(['Canonical_Smiles','std_smiles','std_inchi'], axis='columns', inplace=True)
```

```python
df_activity = df_clean \
    .join(df_molecules, on='Molecule', how='inner') \
    .join(df_molecules_std, on='std_inchi_key', how='inner') \
    .groupby(['std_inchi_key', 'receptor_name']) \
    .aggregate('median') \
    .Standard_Value \
    .to_frame() \
    .reset_index('receptor_name')
```

## Generate fingerprints and scaffolds

```python
std_mols_clean = list(map(MolFromSmiles, df_molecules_std.std_smiles))
```

```python
df_molecules_std['std_morgan_fp'] = [AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=1024).ToBitString() for mol in std_mols_clean]
```

```python
scaffolds = [Chem.Scaffolds.MurckoScaffold.GetScaffoldForMol(mol) for mol in std_mols_clean]
```

```python
df_molecules_std['scaffold_smiles'] = [MolToSmiles(mol) for mol in scaffolds]
```

```python
df_molecules_std['scaffold_morgan_fp'] = [AllChem.GetMorganFingerprintAsBitVect(mol, 2, nBits=1024).ToBitString() for mol in scaffolds]
```

```python
df_clean.to_csv('secondary_data/df_clean.csv')
df_molecules_std.to_csv('secondary_data/df_molecules_std.csv')
df_activity.to_csv('secondary_data/df_activity.csv')
```

```python
def plot_kde(x, num_points=1000, *args, **kwargs):
    density = gaussian_kde(x)
    x = np.linspace(np.min(x), np.max(x), num_points)
    plt.plot(x, density(x), *args, **kwargs)
```

```python
plt.hist(np.log10([x.GetNumAtoms() for x in std_mols_clean]), bins=40)
plt.figure()
plot_kde(np.log10([x.GetNumAtoms() for x in std_mols_clean]))
plt.title('Number of atoms (log10)')
```

```python
plot_kde(np.log10(df_activity.Standard_Value))
plt.title('Standard value (log10)')
```

```python
plt.hist(np.log10(list(map(len,df_molecules_std.std_smiles))), bins=50)
plt.figure()
plot_kde(np.log10(list(map(len,df_molecules_std.std_smiles))))
plt.title('Length of smiles (log10)')
```

```python
plt.scatter(np.log10(list(map(len,df_join.std_smiles))), np.log10(df_join.Standard_Value), alpha=0.1, s=5)
plt.title('Standard value vs length of smiles (log-log)')
```

```python
from itertools import product

def plot_kde_2d(p_x,p_y, num_points=100, bw_method=None, *args, **kwargs):
    density = gaussian_kde(np.c_[p_x, p_y].T, bw_method)
    x,y = np.linspace(np.min(p_x),np.max(p_x), num_points), np.linspace(np.min(p_y), np.max(p_y), num_points)
    z = density(np.asarray(list(product(x,y))).T).reshape(num_points,num_points).T
#     z = density(list(zip(*product(x,y)))).reshape(num_points,num_points).T
    plt.pcolormesh(x,y,z, *args, **kwargs)
```

```python
plot_kde_2d(np.log10(list(map(len,df_join.std_smiles))), np.log10(df_join.Standard_Value), num_points=100, bw_method=0.1, cmap='YlOrRd')
plt.title('Standard value vs length of smiles (log-log)')
```

## Analysis on ERa

```python
df_era = df_join.query('receptor_name == "ERa"')
```

```python
df_era['active'] = df_era.Standard_Value < 100
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import normalize, minmax_scale
from sklearn.linear_model import LinearRegression
from scipy.stats import linregress
```

```python
fps = [np.fromstring(fp, np.byte) - ord('0') for fp in df_era.std_morgan_fp]
scaffold_fps = [np.fromstring(fp, np.byte) - ord('0') for fp in df_era.scaffold_morgan_fp]
```

```python
log_ic50 = np.log10(df_era.Standard_Value)
```

```python
from matplotlib import cm

def make_colormap(x, colormap=cm.plasma):
    return colormap(np.int64(minmax_scale(x)*255))
```

```python
pca = PCA()
pc = pca.fit_transform(fps)
pca_scaffold = PCA()
pc_scaffold = pca_scaffold.fit_transform(scaffold_fps)
```

```python
def plot_scatter_colormap(x,y,z):
    plt.figure()
    plt.scatter(x,y, c=make_colormap(z), alpha=0.25)
    plt.figure()
    plt.axis('off')
    plt.imshow(np.expand_dims(np.asarray(cm.plasma.colors),0),aspect=10)
```

```python
plot_scatter_colormap(pc[:,0], pc[:,1],log_ic50.values)
```

```python
plot_scatter_colormap(pc_scaffold[:,0], pc_scaffold[:,1],log_ic50.values)
```

```python
def plot_regression(x,y):
    lr = linregress(x,y)
    plt.scatter(x, y, alpha=0.25)
    min_x = np.min(x)
    max_x = np.max(x)
    line_x = np.asarray([min_x, max_x])
    line_y = lr.intercept + lr.slope*line_x
    plt.plot(line_x, line_y, '--', c='black')
    plt.title('r=%.3f, p=%e, stderr=%.3f' % (lr.rvalue, lr.pvalue, lr.stderr))
```

```python
for i in range(4):
    plt.figure()
    plot_regression(pc[:,i], log_ic50)
```

```python
for i in range(4):
    plt.figure()
    plot_regression(pc_scaffold[:,i], log_ic50)
```

```python
lr = LinearRegression()
lr.fit(fps, log_ic50)
r2 = lr.score(fps, log_ic50)
np.sqrt(r2)
```

```python
lr = LinearRegression()
lr.fit(scaffold_fps, log_ic50)
r2 = lr.score(scaffold_fps, log_ic50)
np.sqrt(r2)
```

```python
rs = []
for nc in tqdm_notebook(range(1,513)):
    lr = LinearRegression()
    lr.fit(pc[:,:nc], log_ic50)
    r2 = lr.score(pc[:,:nc], log_ic50)
    rs.append(np.sqrt(r2))
```

```python
rs_scaffold = []
for nc in tqdm_notebook(range(1,513)):
    lr = LinearRegression()
    lr.fit(pc_scaffold[:,:nc], log_ic50)
    r2 = lr.score(pc_scaffold[:,:nc], log_ic50)
    rs_scaffold.append(np.sqrt(r2))
```

```python
plt.plot(rs)
plt.plot(rs_scaffold)
```
